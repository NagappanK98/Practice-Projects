# Mushroom Project

Here we are going to analyze whether the mushrooms are edible or poisonous

# Importing required Libraries

```python
import warnings
warnings.simplefilter("ignore")
```

# Loading the Dataset

```python
import pandas as pd
df = pd.read_csv("Mushroom.csv")         # mapped the column names as given in the dataset description
df.head()
```

```python
df.columns
```

# Data Preparation

```python
df.describe()
```

```python
# converting the categorical data to numerics using label encoder

from sklearn.preprocessing import LabelEncoder


labelencoder=LabelEncoder()
for i in df.columns:
    df[i] = labelencoder.fit_transform(df[i])
```

```python
df.describe()
```

Here we can see there are some null values in the entire dataset

Mean and median are also dont have much difference.



```python
df=df.drop(["veil-type"],axis=1)         # removing since its 0
```

```python
import matplotlib.pyplot as plt


plt.figure()
pd.Series(df['class']).value_counts().sort_index().plot(kind = 'bar')
plt.ylabel("Count")
plt.xlabel("class")
plt.title('Total no of poisonous(1) and edible(0) mushrooms');
```

```python
import seaborn as sns

plt.figure(figsize=(18,16))
sns.heatmap(df.corr(),linewidths=.1, annot=True)
```

Here we observe that gill-color is negatively correlated

```python
df[['class', 'gill-color']].groupby(['gill-color'], as_index=False).mean().sort_values(by='class', ascending=False)
```

```python
new_var=df[['class', 'gill-color']]
new_var=new_var[new_var['gill-color']<=3.5]
sns.factorplot('class', col='gill-color', data=new_var, kind='count', size=2.5, aspect=.8, col_wrap=4);
```

```python
new_var=df[['class', 'gill-color']]
new_var=new_var[new_var['gill-color']>3.5]

sns.factorplot('class', col='gill-color', data=new_var, kind='count', size=2.5, aspect=.8, col_wrap=4);
```

# Machine Learning Model

```python
X=df.drop(['class'], axis=1)
Y=df['class']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.1)
```

# Decision tree classifier

```python
from sklearn.tree import DecisionTreeClassifier


a = DecisionTreeClassifier()
b = a.fit(X_train, Y_train)
```

```python
import numpy as np

features_list = X.columns.values
feature_importance = a.feature_importances_
sorted_idx = np.argsort(feature_importance)

plt.figure(figsize=(5,7))
plt.barh(range(len(sorted_idx)), feature_importance[sorted_idx], align='center')
plt.yticks(range(len(sorted_idx)), features_list[sorted_idx])
plt.xlabel('Importance')
plt.title('Feature importances')
plt.draw()
plt.show()
```

```python
y_pred=a.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_curve, auc, roc_curve


print("Decision Tree Classifier result is", classification_report(Y_test, y_pred))
```

```python
cfm=confusion_matrix(Y_test, y_pred)

sns.heatmap(cfm, annot = True,  linewidths=.5, cbar =None)
plt.title('Decision Tree Classifier confusion matrix')
plt.ylabel('True label')
plt.xlabel('Predicted label');
```

From this we can see decision tree classifier provides 100% accuracy

# Gaussian Naive Bayes (GaussianNB)

```python
from sklearn.naive_bayes import GaussianNB

c = GaussianNB()
d = c.fit(X_train, Y_train)
```

```python
y_pred_GNB=c.predict(X_test)
```

```python
cfm=confusion_matrix(Y_test, y_pred_GNB)
```

```python
sns.heatmap(cfm, annot = True,  linewidths=.5, cbar =None)
plt.title('Gaussian Naive Bayes confusion matrix')
plt.ylabel('True label')
plt.xlabel('Predicted label')
```

```python
print("Test data using Gaussian Naive Bayes method \n", classification_report(Y_test, y_pred_GNB))
```
